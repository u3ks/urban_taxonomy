```python
def process_wfs(region_name, url, params, count):
    params = params.copy()
    i = 0
    while True:
        
        print(i)
        params['count'] = count,
        params['startIndex'] = i,

        # Make the request
        response = requests.get(url, params=params)
    
        if response.status_code != 200:
            break
        
        with open(f"../data/{region_name}/buildings_response_{i}.xml", "wb") as file:
            file.write(response.content)
```
