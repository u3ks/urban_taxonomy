```python
def process_wfs(region_name, url, params, count):
    params = params.copy()
    i = 0
    while True:
        
        print(i)
        params['count'] = count,
        params['startIndex'] = i,

        # Make the request
        response = requests.get(url, params=params)
    
        if response.status_code != 200:
            break
        
        with open(f"../data/{region_name}/buildings_response_{i}.xml", "wb") as file:
            file.write(response.content)
```

```python
import requests 
def download_url(url, save_path, chunk_size=128):
    r = requests.get(url, stream=True)
    with open(save_path, 'wb') as fd:
        for chunk in r.iter_content(chunk_size=chunk_size):
            fd.write(chunk)

download_url('https://daten-hamburg.de/inspire/hh_inspire_gebaeude_2d_alkis/INSPIRE_HH_Gebaeude_2D_ALKIS_2020-07-15.zip', '../data/hamburg.zip')
```
